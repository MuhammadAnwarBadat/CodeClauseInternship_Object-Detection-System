## **Detection of objects in images using pre-trained Yolov8m model**

```python
import os
from ultralytics import YOLO

# Load pre-trained YOLOv8 model
model = YOLO('/content/drive/MyDrive/Code-Clause-Projects/yolov8m.pt')

# Folder to save output videos
output_folder = "/content/drive/MyDrive/Code-Clause-Projects/output-images/"
os.makedirs(output_folder, exist_ok=True)  # Ensure folder exists

# Path to input video
image_path = "/content/drive/MyDrive/Code-Clause-Projects/testing-images-yolo/image-04.png"

# Run object detection and save the output in the specified folder
results = model.predict(source=image_path, save=True, project=output_folder, name="yolo_output")

print(f"Processed image saved in: {output_folder}/yolo_output/")
```

## **Detection of objects in videos using pre-trained Yolov8m model**

```python
import os
from ultralytics import YOLO

# Load pre-trained YOLOv8 model
model = YOLO('/content/drive/MyDrive/Code-Clause-Projects/yolov8m.pt')

# Folder to save output videos
output_folder = "/content/drive/MyDrive/Code-Clause-Projects/output-videos/"
os.makedirs(output_folder, exist_ok=True)  # Ensure folder exists

# Path to input video
video_path = "/content/drive/MyDrive/Code-Clause-Projects/testing-videos-yolo/video-02.mp4"

# Run object detection and save the output in the specified folder
results = model.predict(source=video_path, save=True, project=output_folder, name="yolo_output")

print(f"Processed video saved in: {output_folder}/yolo_output/")
```
